## Part 3:  Data PreProcessing

```python
# ==========================
# 📌 3. Add Cardiac Indicators
# ==========================

# Age Groups
def age_group(age):
    if age < 40: return "Young"
    elif age < 60: return "Middle"
    else: return "Old"

data["age_group"] = data["age"].apply(age_group)

# Blood Pressure Category
def bp_category(bp):
    if bp < 120: return "Normal"
    elif bp < 130: return "Elevated"
    elif bp < 140: return "Hypertension Stage 1"
    else: return "Hypertension Stage 2"

data["bp_category"] = data["trestbps"].apply(bp_category)

# Cholesterol Risk
def chol_risk(chol):
    if chol < 200: return "Desirable"
    elif chol < 240: return "Borderline"
    else: return "High"

data["chol_risk"] = data["chol"].apply(chol_risk)

# Heart Rate Reserve (HRR)
data["HRR"] = (220 - data["age"]) - data["thalach"]

# Oldpeak Risk
def oldpeak_risk(op):
    if op < 1: return "Low"
    elif op < 2: return "Moderate"
    else: return "High"

data["oldpeak_risk"] = data["oldpeak"].apply(oldpeak_risk)

# Composite Risk Score
data["risk_score"] = (
    (data["age"]/100) + 
    (data["trestbps"]/200) + 
    (data["chol"]/300) + 
    (data["oldpeak"]/5)
)

# Encode categorical indicators
for col in ["age_group", "bp_category", "chol_risk", "oldpeak_risk"]:
    data[col] = LabelEncoder().fit_transform(data[col])
```

##  Feature Engineering: Adding Clinical Context

Raw medical data is valuable, but Machine Learning models perform significantly better when we provide them with **domain-specific insights**. In this section, we apply medical thresholds to transform raw numbers into meaningful clinical categories. 

This helps the model recognize critical medical patterns faster and more accurately.

**Here is what we are adding:**
* **`age_group`**: Categorizes patients into "Young", "Middle", and "Old" to capture age-related risk trends.
* **`bp_category`**: Maps resting blood pressure (`trestbps`) to standard medical clinical stages (Normal, Elevated, Hypertension).
* **`chol_risk`**: Classifies cholesterol levels into "Desirable", "Borderline", or "High" risk based on medical standards.
* **`HRR` (Heart Rate Reserve)**: A widely used medical formula `(220 - Age - Max Heart Rate)` that indicates cardiovascular fitness and heart capacity.
* **`oldpeak_risk`**: Categorizes the ST depression value into severity levels (Low, Moderate, High).
* **`risk_score`**: A composite, weighted numerical score that aggregates multiple risk factors (Age, BP, Cholesterol, Oldpeak) into a single indicator of overall heart health.

*(Note: Finally, we use `LabelEncoder` to convert these newly created text categories back into numbers so our Machine Learning algorithms can process them effectively).*

## Part 4 : Modeling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
                             RocCurveDisplay, PrecisionRecallDisplay, ConfusionMatrixDisplay)
import mlflow
import mlflow.sklearn

# ==========================================
# 1. Data Preparation & Scaling 
# ==========================================
x = data.drop(["condition", "condition_label", "sex_label"], axis=1, errors='ignore')
y = data["condition"]
feature_names = x.columns.tolist()

scaler = StandardScaler()
x_scaled = pd.DataFrame(scaler.fit_transform(x), columns=feature_names)
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42, stratify=y)

# ==========================================
# 2. Define Models and Hyperparameter Grids
# ==========================================

models_and_parameters = {
    "RandomForest": {
        "model": RandomForestClassifier(random_state=42, class_weight="balanced"),
        "params": {
            "n_estimators": [100, 300, 500],
            "max_depth": [None, 5, 10, 15],
            "min_samples_split": [2, 5, 10],
            "min_samples_leaf": [1, 2, 5]
        }
    },
    "GradientBoosting": {
        "model": GradientBoostingClassifier(random_state=42),
        "params": {
            "n_estimators": [100, 200],
            "learning_rate": [0.01, 0.1, 0.2],
            "max_depth": [3, 5, 7]
        }
    },
    "LogisticRegression": {
        "model": LogisticRegression(random_state=42, max_iter=9000),
        "params": {
            "C": [0.01, 0.1, 1.0, 10.0], 
            "penalty": ["l2"]
        }
    }
}

# ==========================================
# 3. MLflow Tracking & GridSearchCV
# ==========================================
mlflow.set_tracking_uri("http://127.0.0.1:5000")
mlflow.set_experiment("heart_disease_tuned_experiment")

# استراتيجية التقسيم للـ GridSearch
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, config in models_and_parameters.items():
    with mlflow.start_run(run_name=f"{name}_Tuned"):
        print(f"🔄 Tuning and Training {name}...")
        
        # --- 1. Grid Search (Hyperparameter Tuning) ---
        grid_search = GridSearchCV(
            estimator=config["model"],
            param_grid=config["params"],
            cv=cv_strategy,
            scoring='accuracy',
            n_jobs=-1 # استخدام كل طاقة البروسيسور عشان يخلص أسرع
        )
        
        # هندرب الـ GridSearch عشان يدور على أحسن موديل
        grid_search.fit(x_train, y_train)
        
        # هناخد أحسن موديل هو لقاه
        best_model = grid_search.best_estimator_
        best_cv_score = grid_search.best_score_
        
        # --- 2. Predictions & Metrics using the BEST model ---
        y_pred = best_model.predict(x_test)

        try:
            y_prob = best_model.predict_proba(x_test)
            if y_prob.shape[1] == 2:
                auc = roc_auc_score(y_test, y_prob[:, 1])  
            else:
                auc = roc_auc_score(y_test, y_prob, multi_class="ovr", average="weighted")
        except Exception:
            auc = np.nan 

        accuracy  = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average="weighted")
        recall    = recall_score(y_test, y_pred, average="weighted")
        f1        = f1_score(y_test, y_pred, average="weighted")

        print(f"✅ {name} | Best CV Acc: {best_cv_score:.3f} | Test Acc: {accuracy:.3f} | AUC: {auc:.3f}")
        print(f"   Best Params: {grid_search.best_params_}\n")

        # --- 3. Log to MLflow ---
        mlflow.log_param("model_name", name)
        mlflow.log_params(grid_search.best_params_) 
        
        mlflow.log_metric("best_cv_accuracy", best_cv_score)
        mlflow.log_metric("accuracy", accuracy)
        mlflow.log_metric("precision", precision)
        mlflow.log_metric("recall", recall)
        mlflow.log_metric("f1_score", f1)
        mlflow.log_metric("auc", auc)

        mlflow.sklearn.log_model(best_model, artifact_path="model")

        # --- 4. Plots ---
        try:
            RocCurveDisplay.from_estimator(best_model, x_test, y_test)
            plt.title(f"ROC Curve: {name} (Tuned)")
            plt.savefig(f"roc_{name}.png")
            mlflow.log_artifact(f"roc_{name}.png")
            plt.close()
            
            ConfusionMatrixDisplay.from_estimator(best_model, x_test, y_test)
            plt.title(f"Confusion Matrix: {name} (Tuned)")
            plt.savefig(f"cm_{name}.png")
            mlflow.log_artifact(f"cm_{name}.png")
            plt.close()
        except Exception:
            pass
```
